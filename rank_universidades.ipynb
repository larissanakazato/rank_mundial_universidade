Importando arquivo cwurData.csv

```python
import csv
import pandas as pd
df_rkuniversidades = pd.read_csv('cwurData.csv')
```

```python
df_rkuniversidades['year'].unique()
```

```python
#Seleciona apenas as universidades que estão no ranking em 2015
df_rk2015=df_rkuniversidades.where(df_rkuniversidades['year']==2015)
```

```python
df_rk2015.set_index(['institution','country'], inplace=True)
```

```python
#Quantidade de universidade por país no ranking em 2015
for country,frame in df_rk2015.groupby(level='country'):
  print(f'No país {country} possui {len(frame)} universidade(s)')
```

```python
df_rk2015.dropna()
```

```python
#País com maior quantidade de universidade e com menor quantidade de universidade no ranking em 2015
qtdMaior=0
qtdMenor=1000
paisMenor=''
paisMaior=''
for country,frame in df_rk2015.groupby(level='country'):
  qtd=len(frame)
  if qtdMaior<qtd:
    paisMaior=country
    qtdMaior=qtd

  if qtdMenor>qtd:
    paisMenor=country
    qtdMenor=qtd

print(f'O país {paisMenor} com a menor quantidade de universidades participantes - {qtdMenor} universidade(s)')
print(f'O país {paisMaior} com a maior quantidade de universidades participantes - {qtdMaior} universidade(s)')
```

```python
def mediaUniversidade(linha):
  dados = linha[['score']]
  return pd.Series({
                  "Media Score":dados.mean()
                   })
```

```python
resultado=df_rk2015.apply(mediaUniversidade,axis=1)
```

```python
resultado
```

```python
for country,frame in resultado.groupby(level='country'):
  print(f'País: {country} - {frame.mean()}')
```

```python
maior=0
menor=100
paisMaior=''
paisMenor=''
for country,frame in resultado.groupby(level='country'):
  media=float(frame.mean())
  if maior<media:
    paisMaior=country
    maior=media
  if menor>media:
    paisMenor=country
    menor=media

print(f'País com maior média de Score é {paisMaior} - {maior}')
print(f'País com menor média de Score é {paisMenor} - {menor}')
```

Apesar de os Estados Unidos estarem em 1º lugar no ranking de 2015, o país que obteve a melhor média de Score foi Singapura.
